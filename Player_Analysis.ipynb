### Defensive Statistics
Caused turnovers is one of the most common ways to judge a defensive players performance
Groundballs are a key metric for a team - normally the team with the most ground balls wins the game as more ground balls means more possesion and more chance to score.

```python
dfCausedTurnovers = dfST.sort_values(by="caused_turnovers", ascending=False).head(10)
dfCausedTurnovers
```

```python
chart = sns.barplot(x=dfCausedTurnovers.last_name, y=dfCausedTurnovers.caused_turnovers)
chart.set_xticklabels(chart.get_xticklabels(),rotation=-45)
plt.xlabel("Player (Last Name)")
plt.ylabel("Caused Turnovers")
plt.title("2019 Premier Lacrosse League (PLL) Caused Turnover Leaders")
plt.savefig("img/Caused-Turnovers-per-game.svg", dpi=300, pad_inches=0, bbox_inches ='tight')
plt.show()
```
